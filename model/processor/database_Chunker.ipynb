```python
import chromadb
from transformers import AutoTokenizer, AutoModel
from chromadb.config import Settings
import torch
import numpy as np
from tqdm import tqdm
import os

# Load the pre-trained model and tokenizer
model_name = "HooshvareLab/bert-base-parsbert-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

persist_directory = "chromadb_collections/"
os.makedirs(persist_directory, exist_ok=True)

# Initialize the persistent client
client = chromadb.PersistentClient(path=persist_directory)

# Load or create the collection
collection = client.get_or_create_collection(name="legal_cases")
```

```python
# Move the model to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Helper function to embed a chunk of text
def embed_text_batch(text_list):
    inputs = tokenizer(text_list, return_tensors="pt", truncation=True, padding=True, max_length=512)
    inputs = {key: value.to(device) for key, value in inputs.items()}  # Move inputs to GPU
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # Move result back to CPU and convert to NumPy array
```

```python
batch_size = 10 # Number of cases to process in parallel

# Process each case and store its averaged embedding in ChromaDB
case_embeddings = []
case_metadatas = []
case_ids = []
case_documents = []

for idx in tqdm(range(0, len(df), batch_size)):
    batch_case_texts = df["case_text"][idx:idx + batch_size]
    batch_case_indices = df.index[idx:idx + batch_size]  # Get the indices of the current batch

    for batch_idx, (case_text, case_idx) in enumerate(zip(batch_case_texts, batch_case_indices)):
        # Use the actual index from the DataFrame as part of the case_id
        case_id = f"case_{case_idx}"

        # Check if the case has already been processed
        existing_ids = collection.get(ids=[case_id])
        if existing_ids['ids']:  # Check if the list is not empty
            print(f"Skipping already processed case: {case_id}")
            continue

        words = case_text.split()
        chunk_size = 250
        overlap = 50
        chunks = []

        # Create chunks of words
        chunk_texts = [" ".join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size - overlap)]

        # Embed the chunks in batch
        chunk_embeddings = embed_text_batch(chunk_texts)

        # if idx == 0 and batch_idx == 0:
        #     print(f"Embedding vector dimension: {chunk_embeddings.shape[1]}")

        # Average the chunk embeddings to get the case-level embedding
        case_embedding = np.mean(chunk_embeddings, axis=0)
        case_embedding_list = case_embedding.tolist()

        # Collect embeddings, metadata, and documents
        case_embeddings.append(case_embedding_list)
        case_metadatas.append({"case_id": case_id, "case_text": case_text})
        case_ids.append(case_id)
        case_documents.append(case_text)

    if case_ids:
        # print(f"Adding {len(case_ids)} cases to the collection.")
        collection.add(
            ids=case_ids,
            embeddings=case_embeddings,
            metadatas=case_metadatas,
            documents=case_documents
        )

    # Clear lists for the next batch
    case_embeddings.clear()
    case_metadatas.clear()
    case_ids.clear()
    case_documents.clear()
```

```python
from hazm import *

normalizer = Normalizer()

# embed a single query
def embed_single_text(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    inputs = {key: value.to(device) for key, value in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()

# query the most relevant cases
def retrieve_relevant_cases(query_text, top_k=5):
    normalized_query_text = normalizer.normalize(query_text)
    query_embedding = embed_single_text(normalized_query_text)
    query_embedding_list = query_embedding.tolist()
    results = collection.query(
        query_embeddings=[query_embedding_list],
        n_results=top_k
    )
    return results
```

```python
from chromadb import Documents, EmbeddingFunction, Embeddings
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-mpnet-base-v2')
model
```

```python

class MyEmbeddingFunction(EmbeddingFunction):
    def __call__(self, input: Documents) -> Embeddings:
        sentences = input
        embeddings = model.encode(sentences)

        embeddings_as_list = [embedding.tolist() for embedding in embeddings]

        return embeddings_as_list
```
